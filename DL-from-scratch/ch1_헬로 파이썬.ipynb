## 1.4 파이썬 스크립트 파일

### 1.4.2 클래스

```python
class Man:
    def __init__(self, name):
        self.name = name
        print("Initialized")
    
    def hello(self):
        print("Hello "+ self.name + "!")
    
    def bye(self):
        print("Bye "+self.name+"!")
```

```python
m = Man("jihyun")
m.hello()
m.bye()
```

## 1.5 넘파이
- 수치 계산용 외부 라이브러리, 넘파이의 배열 클래스인 numpy.array에 배열(행렬)을 다루기 편한 메서드가 많음.

```python
import numpy as np
```

```python
# 1.5.2 배열 생성 -> 넘파이가 제공하는 특수한 형태의 배열(numpy.ndarray)로 반환
x = np.array([1.0,2.0])
print(type(x))

# 1.5.3 넘파이 배열로 산술 연산
y = np.array([2.0,3.0])
print(x+y)
print(y/2)
```

```python
# 1.5.4 N차원 배열 - 1차원: 벡터, 2차원: 행렬, 벡터와 행렬을 일반화: 텐서
a = np.array([[1,2,3],[4,5,6]])
print(a.shape)
print(a.dtype)
```

```python
# 1.5.5 브로드캐스트: 스칼라/1차원배열이 a 행렬의 shape으로 변형된 후 연산이 이루어지는 기능
a = np.array([[1,2],[3,4]])
b = np.array([10,20])   # [[10,20],[10,20]]으로 확대됨
a*b
```

```python
# 1.5.6 원소 접근
X = np.array([[12,2],[42,3],[2,5]])
print(X)
X = X.flatten()     # X를 1차원 배열로 평탄화
print(X)
print(X[[0,2,4]])   # i=0,2,4 원소만 얻기
print(X>10)         # bool 배열 반환
print(X[X>10])
```

## 1.6 matplotlib
- pyplot 모듈 사용

```python
import matplotlib.pyplot as plt
```

```python
## 1.6.2 pyplot의 기능
x = np.arange(0,6,0.1)  # 0부터 6미만 까지 0.1 간격으로 배열 생성
y1 = np.sin(x)
y2 = np.cos(x)

# 그래프 그리기
plt.plot(x,y1,label='sin')  # label=데이터 이름
plt.plot(x,y2,linestyle='-.',label='cos')   
plt.xlabel("x") # x축 이름
plt.ylabel("y") # y축 이름
plt.title("sin & cos")    # 그래프 제목
plt.legend()    # 데이터의 label 표시
plt.show()
```
